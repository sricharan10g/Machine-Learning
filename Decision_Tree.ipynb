```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop(["Name", "PassengerId", "Survived", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis="columns")
target = df["Survived"]
inputs.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
n_Pclass = LabelEncoder()
n_Sex = LabelEncoder()
n_Age = LabelEncoder()
n_Fare = LabelEncoder()
```

```python
inputs['new_Pclass'] = n_Pclass.fit_transform(inputs['Pclass'])
inputs['new_Sex'] = n_Pclass.fit_transform(inputs['Sex'])
inputs['new_Age'] = n_Pclass.fit_transform(inputs['Age'])
inputs['new_Fare'] = n_Pclass.fit_transform(inputs['Fare'])
inputs.head()
```

```python
n_inputs = inputs.drop(['Pclass', 'Sex', 'Age', 'Fare'], axis= 'columns')
n_inputs.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(n_inputs,target)
```

```python
model.score(n_inputs, target)
```

```python
model.predict([[2, 1, 7, 7]])
```
